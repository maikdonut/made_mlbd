Take data from here: https://www.kaggle.com/datasets/thorgodofthunder/tvradionewspaperadvertising  
Split this dataset in proportion 80/20 to train and test data 

Data consists of the advertising budgets for three media (TV, radio
and newspapers) and the overall sales in 200 different markets

```scala
import $ivy.`org.scalanlp:breeze_2.13:2.1.0`
```

```scala
import breeze.linalg.{DenseMatrix, DenseVector, csvread, csvwrite, sum}
import java.io.File
```

```scala
val train_data = csvread(file = new File("data/train.csv"), separator = ',', skipLines = 1)
```

```scala
// train_data shape
(train_data.rows, train_data.cols)
```

```scala
// split on features and target
val X_train = train_data(::, 0 until train_data.cols - 1).toDenseMatrix
val y_train = train_data(::, -1).toDenseVector
```

Since the data has a different scale, normalize it by StandardScaler (standardize features by removing the mean and scaling to unit variance)

```scala
import breeze.stats.{mean, variance}

val X_train_scaled = DenseMatrix.zeros[Double](X_train.rows, X_train.cols + 1)
val mean_values = DenseVector.zeros[Double](X_train.cols)
val sd_values = DenseVector.zeros[Double](X_train.cols)
X_train_scaled(::, 0) := DenseVector.fill(X_train.rows, 1.0) // bias

for{
  i <- 1 to X_train.cols
}{
  val m = mean(X_train(::, i - 1))
  val sd = math.pow(variance(X_train(::, i - 1)), 0.5)
  mean_values(i - 1) = m
  sd_values(i - 1) = sd
  X_train_scaled(::, i) := (X_train(::, i - 1) - m)/sd
}
```

We will train Linear regression model using gradient descent. Choose MSE as a cost function.

$$
\nabla{MSE} = \frac{2}{N}X^{T}(Xw - Y)
$$

```scala
def gradient_descent(X: DenseMatrix[Double], y: DenseVector[Double], alpha: Double = 0.01, iterations: Int = 100):DenseVector[Double] = {
  val n = X.cols
  val w = DenseVector.zeros[Double](n) // initialize weights with 0
  for{
    i <- 1 to iterations
  }{
    val hypothesis = X*w
    val loss = hypothesis - y
    val grad = X.t*loss * (2.0/n)
    w := w - alpha * grad
  } 
  w
}
```

```scala
val weights = gradient_descent(X_train_scaled, y_train)
```

Now we can make a prediction and evaluate our model.  
Firstly, load test data and prepare it for evaluation.

```scala
val test_data = csvread(file = new File("data/test.csv"), separator = ',', skipLines = 1)
```

```scala
// split on features and target
val X_test = test_data(::, 0 until test_data.cols - 1).toDenseMatrix
val y_test = test_data(::, -1).toDenseVector
```

Transform test data

```scala
val X_test_scaled = DenseMatrix.zeros[Double](X_test.rows, X_test.cols + 1)
X_test_scaled(::, 0) := DenseVector.fill(X_test.rows, 1.0) // bias

for{
  i <- 1 to X_test.cols
}{
  X_test_scaled(::, i) := (X_test(::, i - 1) - mean_values(i - 1))/sd_values(i - 1)
}
```

Make a prediction

```scala
val prediction = X_test_scaled * weights
```

Evaluate it by MSE and MAE

```scala
val mse_error = sum((y_test - prediction).map(x => x * x)) / y_test.length
val mae_error = sum((y_test - prediction).map(x => math.abs(x))) / y_test.length
```

Write our prediction to file prediction.csv

```scala
csvwrite(file = new File("results/prediction.csv"), mat = prediction.toDenseMatrix.t, separator = ',')
```
